```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Admission_Predict_Ver1.1.csv')
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

## Корреляционный анализ

Колонка серийного номера не понадобится для построения кореляционноой матрицы

```python
df = df.drop(columns=["Serial No."])
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title("Correlation Matrix")
plt.show()
```

## Скрепичная диаграммма

```python
plt.figure(figsize=(8, 6))
sns.violinplot(x="University Rating", y="GRE Score", data=df)
plt.title("График распределения баллов GRE по рейтингу университетов")
plt.show()
```
